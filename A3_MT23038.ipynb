```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tqdm import tqdm
from sklearn.model_selection import KFold

# Transpose the ratings DataFrame to get item-item matrix
ratings_df_transpose = ratings_df.T

# Use the already normalized ratings DataFrame
ratings_normalized_df = ratings_df_transpose 

# Number of nearest neighbors to consider
n_values = [10, 20, 30, 40, 50]

# Number of folds for k-fold validation
k = 5

# Function to calculate cosine similarity between items
def cosine_similarity_items(ratings_df):
    num_items = ratings_df.shape[0]
    similarity_matrix = np.zeros((num_items, num_items))
    for i in range(num_items):
        for j in range(num_items):
            vector_i = ratings_df.iloc[i].values
            vector_j = ratings_df.iloc[j].values
            dot_product = np.dot(vector_i, vector_j)
            norm_i = np.linalg.norm(vector_i)
            norm_j = np.linalg.norm(vector_j)
            similarity_matrix[i, j] = dot_product / (norm_i * norm_j)
    np.fill_diagonal(similarity_matrix, 0)  # Set diagonal elements to 0
    return similarity_matrix


# Function to find top N similar items for each item
def top_similar_items(similarity_matrix, n):
    similar_items = np.argsort(similarity_matrix, axis=1)[:, -(n+1):-1]  # Exclude the item itself
    return similar_items


# Function to predict missing ratings
def predict_ratings(train_matrix, similarity_matrix, similar_items):
    predicted_ratings = np.zeros(train_matrix.shape)
    for i in range(train_matrix.shape[0]):
        for j in range(train_matrix.shape[1]):
            if train_matrix[i, j] == 0:  # Predict for missing values
                try:
                    item_similarities = similarity_matrix[j, similar_items[j]]
                    relevant_ratings = train_matrix[:, similar_items[j]]
                    numerator = np.sum(item_similarities * relevant_ratings[i])
                    denominator = np.sum(np.abs(item_similarities))
                    if denominator == 0:
                        predicted_ratings[i, j] = 0
                    else:
                        predicted_ratings[i, j] = numerator / denominator
                except IndexError:
                    print(f"IndexError occurred at i={i}, j={j}")
                    print(f"train_matrix.shape: {train_matrix.shape}")
                    print(f"similarity_matrix.shape: {similarity_matrix.shape}")
                    print(f"similar_items[j]: {similar_items[j]}")
                    raise
    return predicted_ratings

# Function to calculate Mean Absolute Error (MAE)
def calculate_mae(actual_ratings, predicted_ratings):
    mae = np.mean(np.abs(actual_ratings - predicted_ratings))
    return mae

# Perform k-fold cross-validation
kf = KFold(n_splits=k)

mae_values = {}
for n in tqdm(n_values, desc='N Neighbors'):
    mae_values[n] = []
    for train_index, val_index in tqdm(kf.split(ratings_normalized_df), total=k, desc='K-Fold Validation', leave=False):
        # Split data into training and validation sets
        train_data = ratings_normalized_df.iloc[train_index]
        val_data = ratings_normalized_df.iloc[val_index]

        # Calculate similarity matrix and top similar items for training set
        similarity_matrix = cosine_similarity_items(train_data)
        similar_items = top_similar_items(similarity_matrix, n)

        # Predict missing ratings
        predicted_ratings = predict_ratings(train_data.values, similarity_matrix, similar_items)

        # Get actual ratings from validation set
        actual_ratings = val_data.values

        # Calculate MAE
        mae = calculate_mae(actual_ratings, predicted_ratings)
        mae_values[n].append(mae)

# Plot results
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
for n, mae_list in mae_values.items():
    plt.plot(range(1, k + 1), mae_list, label=f'{n} Neighbors')
plt.title('MAE vs K-Fold for Different Number of Neighors (Item-Item)')
plt.xlabel('Fold')
plt.ylabel('Mean Absolute Error (MAE)')
plt.legend()
plt.grid(True)
plt.xticks(range(1, k + 1))
plt.show()
```
